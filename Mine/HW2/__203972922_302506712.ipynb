````python
import random
from collections import defaultdict
import difflib
import sys

_MAXIMUM_DISTANCE = 1000
_MAXIMUM_ITERATIONS = 100

def _initialize_centroids(mX, k):
    centroids = []
    mX_np = np.array(mX)
    n_points = np.array(mX_np).shape[0]
    
    # Randomaly select the first centroid.
    centroids.append(mX_np[np.random.randint(mX_np.shape[0]), :])
   
    for center in range(k - 1): 
        dist = []
        for i in range(mX_np.shape[0]):
            point = mX_np[i, :]
            d = sys.maxsize
            for j in range(len(centroids)):
                temp_dist = _distance(point, centroids[j])
                d = min(d, temp_dist)
            dist.append(d)
              
        # Select the point with maximum distance for the next centroid.
        next_centroid = mX_np[np.argmax(dist), :] 
        centroids.append(next_centroid) 

        # Reinitialize the distance array for next centroid.
        dist = np.empty(n_points)
    return centroids

def _initialize_clusters(mX, K):
    centroids = _initialize_centroids(mX, K)
    clusters = defaultdict(dict)
    for i, x in enumerate(centroids):
        clusters[i]['members'] = list()
        clusters[i]['centroid'] = x
    return clusters
   
def _centroids_close_enough(centroids1, centroids2, maximum_diff=1e-0):
    for i in range(len(centroids1)):
        for j in range(len(centroids1[0])):
            if np.abs(centroids1[i][j] - centroids2[i][j]) <= maximum_diff:
                return False
    return True

def _clear_clusters(clusters):
    for cluster in clusters.values():
        cluster['members'] = list()

def _cluster_mX(clusters, mX):
    '''
    Assigns each point to the relevant cluster
    Args:
        clusters_dict - the clusters and their members
        mX   - Input data,         mX.shape = d, N
    Output:
        clusters_dict - the updated dict of clusters and their new members
    '''
    centroids = [cluster['centroid'] for cluster in clusters.values()] 

    max_dist = _MAXIMUM_DISTANCE
    for j, x in enumerate(mX):
        dists = []
        # Find the clostest centroid and append the point to it.
        for i, centroid in enumerate(centroids):
            dists.append(np.sqrt(np.sum([(x[i] - centroid[i])**2 for i in range(len(x))])))

        minimal_centroid_index = dists.index(min(dists))
        clusters[minimal_centroid_index]['members'].append(j)

def _recompute_centroids(clusters, mX):
    '''
    Recomputes the centroids based on the clusters points.
    Args:
        clusters - the clusters.
        mX   - Input data,         mX.shape = d, N
    Output:
        clusters - the clusters with recomputed centroids.
        
    '''   
    for cluster in clusters.values():
        cluster['centroid'] = [np.mean(x) for x in zip(*[mX[i] for i in cluster['members']])]

def _distance(p1, p2):
    return np.sqrt(np.sum((p1 - p2)**2))

def KMeans(mX, K):
    '''
    Apply K-Means
    Args:
        mX   - Input data,         mX.shape = d, N
        K    - Number of clusters, positive integer
    Output:
        vIdx - Index of the cluster each point belongs to, vIdx.shape = N
        mMu  - Clusters' centers,                          mMu.shape  = d, K
        
    '''    
    clusters = _initialize_clusters(mX, K)
    prev_centroids = [cluster['centroid'] for cluster in clusters.values()]
    curr_centroids =  np.zeros_like(prev_centroids)
    
    i = 0 
    while not _centroids_close_enough(prev_centroids, curr_centroids):
        i += 1
        if i > _MAXIMUM_ITERATIONS:
            break
        _clear_clusters(clusters)
        _cluster_mX(clusters, mX)
        _recompute_centroids(clusters, mX)
        prev_centroids = curr_centroids.copy()
        curr_centroids = [cluster['centroid'] for cluster in clusters.values()]
    
    # Map each point to its cluster.
    labels_list = [0]*len(mX)
    for i, cluster in clusters.items():
        for member in cluster['members']:
            labels_list[member] = i
    
    return labels_list
              
````

## 2 GMM
The GMM algorithm aims to maximize the (log) likelihood objective:
$$\arg\max_{\left\{ \left(w_{k},\boldsymbol{\mu}_{k},\boldsymbol{\Sigma}_{k}\right)\right\} _{k=1}^{K}}f=\arg\max_{\left\{ \left(w_{k},\boldsymbol{\mu}_{k},\boldsymbol{\Sigma}_{k}\right)\right\} _{k=1}^{K}}\sum_{i=1}^{N}\log\left(\sum_{k=1}^{K}w_{k}\mathcal{N}_d\left(\boldsymbol{x}_{i};\boldsymbol{\mu}_{k},\boldsymbol{\Sigma}_{k}\right)\right)$$
where:
$$\mathcal{N}_{d}\left(\boldsymbol{x};\boldsymbol{\mu},\boldsymbol{\Sigma}\right)=\frac{1}{\sqrt{\left(2\pi\right)^{d}\left|\boldsymbol{\Sigma}\right|}}\exp\left(-\frac{1}{2}\left(\boldsymbol{x}-\boldsymbol{\mu}\right)^{T}\boldsymbol{\Sigma}^{-1}\left(\boldsymbol{x}-\boldsymbol{\mu}\right)\right)$$
### 2.3 Implementation
Implement the GMM algorithm.
```python
def GMM(mX, K, =1e-6, maxIter=100):
```
**Notes**:
* Stopping criteria:
    * The objective $f$ has changed less than $\epsilon$.
    * Maximum number of iteration.
* The index $s$ of the point $\boldsymbol{x}_i$ is defined by:
$$\boldsymbol{x}_{i}\in\mathcal{D}_{s} \iff s=\arg\max_{k}w_{k}\mathcal{N}_d\left(\boldsymbol{x}_{i},\boldsymbol{\mu}_{k},\boldsymbol{\Sigma}_{k}\right)$$
* Use [broadcasting](https://numpy.org/doc/stable/user/basics.broadcasting.html) and avoid `for` loops in Step II (parameters update). 


````python
def _initialize_mus(k):
    return np.random.uniform(low=-2, high=2, size=(k,))

# w * N(x; mu, sig^2)
def _Pz(x, w, vMu, mSig):
    return w * sp.stats.multivariate_normal.pdf(x, vMu, mSig)

def GMM(mX, K, =1e-6, maxIter=100):
    '''
    Apply GMM
    Args:
        mX      - Input data,                   mX.shape = (d, N)
        K       - Number of clusters,           positive integer
               - Stopping criterion threshold, positive real
        maxIter - Maximum number of iterations, positive integer
    Output:
        vIdx - Index of the cluster each point belongs to, vIdx.shape = (N,)
        vW   - The weight of each Gaussian,                vW.shape   = (K,)
        mMu  - Centers,                                    mMu.shape  = (d, K)
        mSig - Covariance,                                 mSig.shape = (d, d, K)
    '''
    d = len(mX[0])
    w_list = [1/K for i in range(K)]
    mu_list = [_initialize_mus(d) for i in range(K)]
    sigma_list = [np.eye(d) for i in range(K)]
    
    vW     = np.array(w_list)
    mMu   = np.array(mu_list).T
    mSig   = np.dstack((w for w in sigma_list))
        
    N = len(mX)
    ii = 0
    loss = _MAXIMUM_ITERATIONS
    prev_loss = 0
    while ii < maxIter and abs(loss - prev_loss) > : 
        
        # Estimate probabilites.
        Px = list()
        Ni = list()
        for i in range(len(vW)): 
            Px.append(_Pz(mX, vW[i], mMu[:,i], mSig[:,:,i]))
      
        vSum = sum([Px[i] for i in range(len(vW))])
        for i in range(len(vW)):
            Px[i]  = Px[i] / vSum


        # Estimate parameters.
        for i in range(len(vW)):
            Ni.append(np.sum(Px[i]))
            vW[i] = Ni[i]/ N
            mMu[:,i]    = np.sum(Px[i][:,None] * mX, axis=0) / Ni[i]
            mSig[:,:,i] = (Px[i][:,None] * (mX - mMu[:,i])).T @ (mX - mMu[:,i]) / Ni[i]
            
        ii += 1
        prev_loss = loss
        labels_lst = [np.array([Px[j][i] for j in range(K)]).argmax(axis=0) for i in range(len(mX))]
        loss = np.log2(sum([Px[label][i] for i, label in enumerate(labels_lst)]))
            
    return labels_lst, vW, mMu, mSig
````

## 4 DBSCAN
### 4.1 Implementation
Implement the DBSCAN algorithm.
```python
def DBSCAN(mX, Z, r):
```
**Notes**:
* Noise points should have index `-1`.
* Implement an auxiliary function to compute connected components (using BFS or DFS).

````python
####################################################
from collections import defaultdict
import operator

def _distance(point_i, point_j):
    '''
    Computes auclidian distances between two points 
    Args:
        point_i
        point_j    
    Output:
        distance
    '''
    return np.sqrt(np.sum([(point_i[i] - point_j[i])**2   for i in range(len(point_i))]))

def _compute_distances_table(mX, r, Z):
    '''
    Computes distances between the all points, returns core points and their 
    "neighbors" as well.
    Args:
        mX   - Input data,                mX.shape = d, N.
        r    - Neighborhood radius,       positive real.  
    Output:
        full_distances_table - distance between each point
        core_points - core points and their "neighbors"
    '''
    core_points = defaultdict(list)
    full_distances_table = defaultdict(dict)
    
    for i, point_i in enumerate(mX):
        
        for j in range(i+1, len(mX)):
            
            point_j = mX[j]
            dist = _distance(point_i, point_j)

            full_distances_table[i][j] = dist
            full_distances_table[j][i] = dist
            if dist < r:
                core_points[i].append(j)
                core_points[j].append(i)
                
    core_points = {i:points for i, points in core_points.items() if len(points) > Z}

    return core_points, full_distances_table

def _find_root(node,root):
    while node != root[node][0]:
      node = root[node][0]
    return (node,root[node][1])

def _compute_connected_components(G):
    '''
    Computes all connected components using BFS.
    Args:
        G - a graph
    Output:
        connected_components - each component with their members

    '''
    curr_root = {} 
    for node in G.keys():
        curr_root[node] = (node,0)  
    for node_i in G: 
        for node_j in G[node_i]: 
            (root_i, depth_i) = _find_root(node_i, curr_root) 
            (root_j, depth_j) = _find_root(node_j, curr_root) 
            if root_i != root_j: 
                min_root = root_i
                max_root = root_j 
                if  depth_i > depth_j: 
                    min_root = root_j
                    max_root = root_i
                curr_root[max_root] = (max_root,max(curr_root[min_root][1]+1, curr_root[max_root][1]))
                curr_root[min_root] = (curr_root[max_root][0],-1) 
    connected_components = {}
    for node_i in G: 
        if curr_root[node_i][0] == node_i:
            connected_components[node_i] = []
    for node_i in G: 
        connected_components[_find_root(node_i, curr_root)[0]].append(node_i) 
    return connected_components

def _build_G(mX, r, core_points):
    '''
    Builds graph G(C, E) where all points x_i, x_j (of mX) are connected if 
     dist(x_i, x_j) < r. Returns also the boundery points list (all the points 
      within radius r from thier core point).
    Args:
        mX   - Input data,                                   mX.shape = d, N
        r    - Neighborhood radius,                          positive real.
        core_points - points that at least Z points are within radius r from it 
    Output:
        G   - the graph
        boundery_points - distance between each point
    '''
    boundery_points = []
    G = defaultdict(list)
    core_points_list = list(core_points.keys())
    for i, i_core_point in enumerate(core_points_list):
        boundery_points.extend(core_points[i_core_point])
        for j in range(i+1, len(core_points)):
            j_core_point = core_points_list[j]

            if _distance(mX[i_core_point], mX[j_core_point]) < r:
                G[i_core_point].append(j_core_point)
                G[j_core_point].append(i_core_point)
    return G, boundery_points


def DBSCAN(mX, Z, r):
    '''
    Apply DBSCAN
    Args:
        mX   - Input data,                                   mX.shape = (d, N)
        Z    - Number of points required to be a core point, positive integer
        r    - Neighborhood radius,                          positive real.
    Output:
        vIdx - Index of the cluster each point belongs to,   vIdx.shape = (N,)
    '''
    core_points_dict, full_distances_table = _compute_distances_table(mX, r, Z)
    core_points = list(core_points_dict.keys())
    
    G, boundery_points = _build_G(mX, r, core_points_dict)
    boundery_points = set(boundery_points) - set(core_points)

    connected_components = _compute_connected_components(G)
    
    # Create a map from each core point to it's component.
    points_comps = defaultdict(lambda:-1)
    for comp, comp_lst in connected_components.items():
        for point in comp_lst:
            points_comps[point] = comp
    
    # Map boundery points to their closest component.
    for boundery_point in boundery_points:
        min_dist_point = {'closest_point': -1, 'dist': -1}
        for point, point_dist in full_distances_table[boundery_point].items():

            if point in boundery_points or point in core_points:
                    if point_dist < min_dist_point['dist']:
                        min_dist_point = {'closest_point': point, 'dist': point_dist}
        
        relevant_component = points_comps[min_dist_point['closest_point']] 
        points_comps[boundery_point] =  relevant_component      
        
        if relevant_component != -1:
            connected_components[relevant_component].append(boundery_point)
    
    components = [-1] * len(mX)
    i = 0 
    for comp_list in connected_components.values():
        for point in comp_list:
            components[point] = i
        i += 1
    return components, core_points    
####################################################
````
